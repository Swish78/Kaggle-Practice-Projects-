# Importing Libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/kaggle/input/bank-marketing-data/bank-full.csv',sep=';')
```

```python
y = dataset["y"]
dataset.drop("y", axis=1,inplace = True)
```

```python
dataset = dataset.drop(["default","contact","day","month","campaign","previous","poutcome"], axis=1)
```

```python
dataset = dataset.drop(["pdays"], axis=1)
```

# Encoding Categorical columns 

```python
dataset.loan.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset.loan = le.fit_transform(dataset.loan)
```

```python
dataset.housing.nunique()
```

```python
dataset.housing = le.fit_transform(dataset.housing)
```

```python
dataset.head(3)
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_cols = [col for col in dataset.columns if dataset[col].dtype == 'object']
encoder = OneHotEncoder()
one_hot = encoder.fit_transform(dataset[cat_cols])
one_hot_df = pd.DataFrame(one_hot.toarray(), columns=encoder.get_feature_names_out(cat_cols))
dataset = pd.concat([dataset, one_hot_df], axis=1)

dataset.drop(cat_cols, axis=1, inplace=True)
```

```python
# X.drop("y",inplace = True)
X = dataset
X
```

```python
y = le.fit_transform(y)
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Training the Logistic Classification model on the Training set
score - 0.8900807254229791

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

# Training the K-NN Classification model on the Training set
Score - 0.8768107928784695

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

# Training the SVC Classification model on the Training set
Score - 0.8838880902355413

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

# Training the Random Forest classification model on the Training set

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred2 = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred2)
print(cm)
print(accuracy_score(y_test, y_pred2))
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
